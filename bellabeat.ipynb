## Data Exploration and Analysis

After cleaning the dataset and fixing all the necessary formats, I begin the analysis process firstly by grouping both hourly intensities and hourly calories by time to gain insight into periods of high and low intensities and calories.

```r
# grouping hourly intensities and calories by time

int_new <- hourlyintensities %>%
  group_by(time) %>%
  drop_na() %>%
  summarise(mean_total_int = mean(totalintensity))
view(int_new)

calories_new <- hourlycalories %>%
  group_by(time) %>%
  drop_na() %>%
  summarise(mean_total_cal = mean(calories))

```

#### plotting bar charts

Next, I plot a bar chart to help with my visualization process

```r
# bar charts, echo=FALSE}

ggplot(data = int_new, aes(x=time, y=mean_total_int)) +
  geom_histogram(fill = 'blue', stat = 'identity') +
  theme(axis.text.x = element_text(angle = 90)) +
  labs(title="Average Total Intensity vs. Time")

ggplot(data=calories_new, aes(x=time, y=mean_total_cal)) + 
  geom_histogram(fill = 'blue', stat = 'identity') +
  theme(axis.text.x = element_text(angle = 90)) +
  labs(title="Average Total Calories vs. Time")
```

After visualizing the total intensities and calories over time, I observed that highest period for intensities and calories burned are between 5pm and 7pm. The next stage of my analysis is merging and exploring the  sleep and activity table to check for relationships between activities and how it affects sleeping pattern.

```r
# Merging Sleep and activity data set

merged_sleep_activity <- merge(sleep_day, dailyactivity, by=c('id', 'date'))
```

```r
# plotting scatter plot 

ggplot(data = merged_sleep_activity) +
  geom_point(mapping = aes(x=totalsteps, y=calories)) +
  geom_smooth(mapping = aes(x=totalsteps, y=calories)) +
  labs(title = "Total Steps VS Calories")
```

The graph above shows a positive correlation between steps and calories. This means that an increase in steps taken translates to more calories burned.

```r
# scatter plot Total minuts Asleep VS Sedentary Minutes

ggplot(data = merged_sleep_activity) +
  geom_point(mapping = aes(x=totalminutesasleep, y=sedentaryminutes)) +
  geom_smooth(mapping = aes(x=totalminutesasleep, y=sedentaryminutes)) +
  labs(title = "Total minutes Asleep VS Sedentary Minutes")
```

The next plot shows a negative relationship between total minutes asleep and sedentary minutes. What this says is that an increase in idkle time will lead to less sleep time.

Finally, I carry out a summary statistics on the merged sleep and activity dataset

```r
# Summary Statistics

merged_sleep_activity %>%
  select(calories, totalsteps, veryactiveminutes, fairlyactiveminutes, lightlyactiveminutes, sedentaryminutes, totalsleeprecords, totalminutesasleep) %>%
  summary()
```

The result of the above analysis shows  some noteworthy observations. Firstly, our summary shows that the participants spend an average of 12 hours daily on sedentary lifestyles whereby the mean time spent on very intense activities is 25 minutes.  Secondly, the total average daily steps recorded is 8515. Finally, the participants record an avergae daily sleep duration of 7 hours.
